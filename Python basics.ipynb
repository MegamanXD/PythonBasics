# **5. Arithmetic Operator**

```
x = 9
y = 3
```

```
plus = x + y
print("Plus = {result}".format(result = plus))

minus = x - y
print("Minus = {result}".format(result = minus))

product = x * y
print("Product = {result}".format(result = product))

division = x / y    # Always return float
print("Division = {result}".format(result = division))

power = x ** y
print("Power = {result}".format(result = power))
```

```
x = 10
y = 3

floor_division = x // y
print("Floor Division = {result}".format(result = floor_division))

remainder_division = x % y
print("Remainder Division = {result}".format(result = remainder_division))
```

**Order of evaluation**: Brackets >> Exponents >> Division >> Multiplication >> Addition >> Subtraction

# **18. args & kwargs**

```
def func(x):
  def func2():
    print(x)
  return func2      # Return the function instead of a value

print(func(4))
```

```
func(4)()

# This is the same as
x = func(4)     # Get the function and put it into x
x()             # Run the function inside x
```

```
x = [1, 23, 234241, 14345]
print(*x)
```

```
def func(x, y):
  print(x, y)

pairs = [(2, 3), (4, 5)]
for pair in pairs:
  func(*pair)
```

```
func(**{'x': 2, 'y': 5})    # ==> x=2, y=5, then print Values
print(*{'x': 2, 'y': 5})    # Keys
```

```
def func(*args, **kwargs):
  print(args)   # This returns a tuple
  print(kwargs) # This returns a dictionary

func(1,2,3,4,5, one=2, two=1)
```

# **19. Scope & Globals**

```
x = "Minh"

def func(name):
  x = name

print(x)

func("changed")
print(x)
```

```
x = "Minh"

def func(name):
  global x
  x = name

print(x)

func("changed")
print(x)
```

# **22. Map and Filter**

```
x = [1231231,2,41,23,12,312,34,54,33,2,43,12,31,52,45,345,345,2,2312,31,23]

# Create a new list by injecting a lambda function to an existing list
x_plus_2 = map(lambda i : i+2, x)     
x_plus_2 = list(x_plus_2)

print(x)
print(x_plus_2)
```

```
x = [1231231,2,41,23,12,312,34,54,33,2,43,12,31,52,45,345,345,2,2312,31,23]

# Filter a list based on a condition
x_is_even = filter(lambda i : i%2==0, x)     
x_is_even = list(x_is_even)

print(x)
print(x_plus_2)
```
